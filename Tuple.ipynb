```python
#python supports indexing and slicing bcz background data structre is an array
#tuple presevers sequential order
#duplicates are allowed in tuple
# tuple is immutable in nature - cannot be changed
#in data type we have 2 approaches:
#1. mutable- can perform changes directly(add,remove,insert,delete)
#ex. list,set,dict
#2. immutable- cannot perform the change in orignal object,change remain temp.
#eg. tuple,string,range
```

```python
p=[10,20,30]
p
```

```python
#list is mutable changes the original object
#tuple is immutable does not change unless we reassign it
```

```python
p[-1]='aditi'
p
```

```python
#remove 10
p.remove(10)
p
```

```python
#use insert (index,vslue)to add element using index
p.insert(0,'laxmi')
p
```

```python
#lets try to understand immutable concept
#in which user cannot be able to make chages in original object
t=(10,20,30)
t
```

```python
#replace last element
t[-1]='aditi'
t
```

```python
#tuple contains 2 methods
#count and index
#count - checks how many objects are repeated in an object
#index-gives indexing of the objects in the tuple
```

```python
t=(10,20,30)
t.index(30)
```

```python
#counts how many times the no is repeated
t=(10,20,30,10,10)
t.count(10)
```

```python
#tuple packing and unpacking
#packing : combines segregated/seperated elements as a one entity
```

```python
a=100
b=200
c=300
a,b,c
```

```python
#packing variables can be int,float,str anything which gets combined as one entity
nums=a,b,c
nums
```

```python
#unpacking- process of seperating individual object and store it in seperate variable called as nums
num1=100
num2=300
name='aditi'
num1,num2,name=nums
nums
```

```python
num1=10
num2=20
name='aditi'
final=num1,num2,name
final
```
