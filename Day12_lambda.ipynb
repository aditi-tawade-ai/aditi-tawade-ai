# **Higher Order function**

```python
#higher order function- it is a function who need another function as an input
#function(function)
```

```python
# map, filter and reduce are higher order functions
```

```python
#iterable- colection of elements--.str,list,tuple,array,etc
# map(Function,iterable)
k = [20,5,4,60,10]

# apply an operation over each element from an iterable

map(lambda num: num*num,k)
```

```python
#when we get an op in the form of object then we have 2 choices
#choice 1 - type caste
tuple(map(lambda num: num*num,k))
```

```python
#choice 2- use of for loop
out = map(lambda num: num*num,k)
for val in out:
  print(val)
```

```python
names=('gayatri','abdul','snehal','jay')

list(map(lambda n:n.upper(),names))

```

```python
#filter (function,itterable)
#this is conditional function
#the value for which condition is true,will get fetched and operations
#will be performed if required

p = (12,-33,40,-8,99,-12)
#filter -ve numbers

list(filter(lambda num:num<0,p))
```

```python
list(filter(lambda num:num>0,p))
```
