```python
#Q12. find the sum of 1st and last digit
num=int(input('Enter the number:'))
last= num % 10   #gives the last remainder

first=num
while first>=10:   #becoz the 1st single digit is always small than 10 if condition is false we get 1st no.
  first=first//10   # it removes last digits until no. is less than 10
  sum=first + last
print('First digit:',first)
print('Last digit:',last)
print('sum of first and last digit:',sum)
```

```python
#Q22. Find HCF of 2 num
a=int(input('Enter number_1='))
b=int(input('Enter number_2='))
while b!=0:
  temp=b
  b=a%b
  a=temp
print('HCF is:',a)
```

```python
#Q23. Find LCM of 2 nums
a=int(input('Enter number 1= '))
b=int(input('Enter number 2= '))
x=a
y=b
while y!=0:
  temp=y
  y=x%y
  x=temp
  lcm=(a*b) // x
print('LCM is:',lcm)
```

```python
#Q27. Armstrong no. or not
num=(input('Enter a number:'))

power=len(num)
total=0
for n in num:
  total=total+ int(n)**power
if total==int(num):
    print('It is an Armstrong number')
else:
    print('It is not an Armstrong number')
```

```python
#Q28. print all armstrong no. between 1 to n
n=500

for num in range(1,n+1):
  power=len(str(num))
  total=0
  for digit in str(num):
    total=total+int(digit)**power
  if total==int(num):
    print(total,end=',')
```

```python
#Q29. Fibonacci series up to n terms
n=10
pre_num=0
new_num=1

for i in range(n):
  print(pre_num,end=',')
  sum=pre_num+new_num
  pre_num=new_num
  new_num=sum
  print(sum,end=',')
```
