```python
gender=['male','female','female','male']
#convert male to 0 and female to 1

[0 if val=='male'else 1 for val in gender]
```

```python
salary= [24000,12500,45000,65000,11000,10000]
#PS: if salary > than 25k 10% bonas else 20%
[sal+sal*0.1 if sal>25000 else sal+sal*.2 for sal in salary]
```
