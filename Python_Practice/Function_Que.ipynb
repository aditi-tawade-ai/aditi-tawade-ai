```python
#Q4. op of one function is passed as input to other function
def result(score):
  if score < 35:
    return 'Fail'
  else:
    return 'Pass'

def display(status):
 if status=='Pass':
    print('Congrats,you passed the exam!')
 else:
    print('You failed the exam')
mks=int(input('Enter your marks: '))
res=result(mks)
display(res)
```

```python
#Q7. fun to add n no. of values
def add(*num):
  total=0
  for n in num:
    total=total+n
  return total
print('sum:',add(10,20,30,40,50))
```

```python
#Q8. accept students detail using **kwargs
def detail(**info):
  print(info)
detail(Aditi= 'roll no.1',sayee= 'roll no.2',jay= 'roll no.3',arya= 'roll no.4')
```

```python
def detail(**info):
  for key,value in info.items():
    print(key,':',value)
detail(Aditi= 'roll no.1',sayee= 'roll no.2',jay= 'roll no.3',arya= 'roll no.4')
```

```python
#Q11. factorial of num using recusive function
def num(n):
  if n == 0:
    return 1
  return n* num(n-1)
print('Factorial:',num(8))
```

```python
def fib(n):
  if n<=1:
    return n
  return fib(n-1)+fib(n-1)
print('Series:',fib(10))
```

```python
#Q12. Write a Python function with a docstring and type hints to calculate simple interest.
def si(pri:float,rate:float,time:float)  -> float:
  '''inputs for calculating SI'''
  return(pri*rate*time)/100
interest = si(1000,2,3)
print('SI=',interest)

```
