```python
#Q10. count total no. of notes in given amount

num=int(input('Enter a amount :'))
deno=[500,200,100,50,20,10,5,2,1]

for i in deno:   #i goes to all nos. in deno
  notes=num//i   #it checks how many notes(i) can be used in num
  remaining_amt=num%i  #remainder after using the notes
  #new_amt=remaining_amt
  if notes>0:
    print(i,':',notes)
  num=remaining_amt  #updates the number and repeats the process
```

```python
#Q12. check types of triangle

a=int(input('Enter angle a: '))
b=int(input('Enter angle b: '))
c=int(input('Enter angle c: '))

if (a==b==c) and(a+b+c==180):
  print('It is an Equilateral triangle')
elif(a==b or b==c or c==a) and (a+b+c)==180:
  print('It is an Isosceles triangle')
elif(a!=b!=c) and (a+b+c)==180:
  print('It is a scalene triangle')
else:
  print('Not valid')
```

```python
#Q13. calculate profit or loss

selling_cost=int(input('Enter selling price:'))
actual_cost=int(input('Enter actual cost:'))
a=selling_cost
b=actual_cost

if a>b:
  profit=((a-b)/b)*100
  print('Profit of:',int(profit),'%')
elif a<b:
  loss=((b-a)/b)*100
  print('Loss of:',int(loss),'%')
else:
  print('No profit or loss')
```

```python
#Q14. calculate percentage with grade

a=float(input('Enter Physics marks:'))
b=float(input('Enter Chemistry marks:'))
c=float(input('Enter Biology marks:'))
d=float(input('Enter Mathematics marks:'))
e=float(input('Enter Computer marks:'))
f=int(input('Enter total marks'))
total=(a+b+c+d+e)
percentage=(total/f)*100

if percentage>=90:
  print('Percentage you got:',percentage,'% with Grade A')
elif percentage>=80:
  print('Percentage you got:',percentage,'% with Grade B')
elif percentage>=70:
  print('Percentage you got:',percentage,'% with Grade C')
elif percentage>=60:
  print('Percentage you got:',percentage,'% with Grade D')
elif percentage>=40:
  print('Percentage you got:',percentage,'% with Grade E')
elif percentage<40:
  print('Percentage you got:',percentage,'% with Grade Fail')
```
