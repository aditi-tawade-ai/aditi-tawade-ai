```python
#Q1. print 1 to 5 using while loop
i=1
while i<=5:
  print(i)
  i+=1
```

```python
#Q2. print num 10 to 1 using while loop
i=10
while i>0:
  print(i)
  i-=1
```

```python
#Q3. Sum of 1st 10 natural num
i=1
sum=0
while i<=10:
  sum=sum+i
  i+=1
print('Sum:',sum)
```

```python
#Q4.Multiplication table
num=5
i=1
while i<=10:
  print(i,'x',num,'=',i*num)
  i+=1
```

```python
#Q5. password inp until correct
password=''
while password!='1234':
  password=input('Enter password:')
print('Correct password')
```

```python
#Q6.Guess the num game
num=''
while num!='7':
  num=input('Guess again:')
print('Correct guess')
```

```python
#Q7. even num from 2 to 20
i=2
even=[]
while i<=20:
 i%2==0
 even.append(i)
 i+=2
print(even)
```

```python
#Q8. factorials of a num
num=int(input('enter a num:'))
fact=1
while fact<=num:
  fact=fact*num
  num+=1
print(fact)
```

```python
#Q9. menu choices until exit
choice=0
while choice!=4:
  print('choose 1 for pizza')
  print('choose 2 for burger')
  print('choose 3 for hotdog')
  print('choose 4 for exit')
  choice=int(input('Enter your choice:'))
  if choice==1:
    print('You ordered pizza')
  elif choice==2:
    print('You ordered burger')
  elif choice==3:
    print('You ordered hotdog')
  elif choice==4:
    print('Exit')
  else:
    print('Invalid choice')
```

```python
#Q10. ATM pin with limited attempts
pin ='1234'
#op=input('Enter the pin')
attempt=3
while attempt>0:
  op=input('Enter the pin:')
  if op!= pin:
    attempt=attempt-1
    print('Wrong pin try again')
    #print('Attempts left:',attempt)
  else:
    print('Login successful..!!')
    break
if attempt==0:
  print('No attempts left')
```

```python
#Q11. Repeatedly ask for name and exit when typed
data=input('Enter your name:')
while data!='exit':
  print('Hello',data)
  data=input('Enter your name:')
print('Exit')
```

```python
#Q12. total maeks until mark 0
mks=int(input('Enter marks or o for calculating:'))
sum=0
while mks!=0:
  print('Enter marks or 0 for calculating:')
  sum=sum+mks
  mks=int(input())
print('Total marks:',sum)
```
