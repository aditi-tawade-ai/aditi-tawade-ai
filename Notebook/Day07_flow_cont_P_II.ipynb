```python
#assignment : fetch most frequent nos. from list
k=[1,1,2,2,2,3,4,5]
max_val=0
answer=[]

for val in k:
  count=k.count(val)
  if count > max_val:
    max_val= count
    answer=[val]
  elif count==max_val and val not in answer:
    answer.append(val)
print('Most frequent value:',answer)
print('with frequency:',max_val)
```

```python
k=[1,1,2,2,2,3,3,4,5]
for val in k:
  if k.count(val) >1 and val not in answer:
    answer.append(val)
print(answer)
```
