```python
import pandas as pd
```

```python
path = '/content/Salary_Data.csv'
path
```

```python
# load dataset
sal=pd.read_csv(path)
sal[:2]
```

```python
# prepare data
X = sal[['YearsExperience']]   #ML alogrithm always takes ip as 2D
y = sal['Salary']
```

```python
# check relationship od the data
import matplotlib.pyplot as plt

plt.scatter(X,y,color='red',marker='*')
```

```python
# import alogrithm --> scikit learn --> sklearn

from sklearn.linear_model import LinearRegression
```

```python
# as it is a class we need to create a object

model = LinearRegression()
model
```

```python
model
```

```python
# Training stage --> using fit method

model.fit(X,y)
```

```python
# Training means it is calculating the slope n intercept with is
# required for formation of line equation
```

```python
# Testing phase

import warnings
warnings.filterwarnings('ignore')

model.predict([[0]])
```

# **Y = mx+c**

```python
# m = slope
m = model.coef_
m
```

```python
# c = intercept
c = model.intercept_
c
```

```python
test_data=[[0],[4.5],[6],[11]]
```

```python
# since we used new values as experience yrs which are not in data
# so to calculate the salary we used y_pred to store the new data we got
# y containes original data and y pred containes new

y_pred = model.predict(test_data)
y_pred
```

```python
# actual plot
plt.scatter(X,y,color='red',marker='o')

# print line of prediction
plt.plot(test_data,y_pred,color='k')
```

```python
plt.plot(X,y,color='red',marker='o')

# print line of prediction
plt.plot(test_data,y_pred,color='k')
```

```python
model.score(X,y)
```

```python
# since we know the data is LR we can train the data
# for training we need to import the algorithm 1st
# we need to import sklearn from that linearRegression

from sklearn.linear_model import LinearRegression
```

```python
# now linearRegression is a function so we need to give object
# for it

model = LinearRegression()
model
```
