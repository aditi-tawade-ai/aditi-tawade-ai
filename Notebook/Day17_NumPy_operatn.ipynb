# **Numpy Operarions**

```python
#create an numpy array of 4x4 zeros
import numpy as np
np.zeros((4,4))
```

```python
np.full((4,4),1)  #here we can gie any value we want
#no string because it gives dtype as unique
```

```python
np.full((4,4),0,dtype=bool)
```

```python
np.full((4,4),-11,dtype=float)
```

```python
#create a numpy array of 4x4 with diagonal elements as true and non diagonal elements as false
np.identity(4)  #square matrix-2D -->[[]]
```

```python
#in python except 0 & none,[],() everything is true
np.identity(4,dtype=bool)
```

```python
bool('adi')
```

```python
#create a numpy array of 4x4 with diagonal elements as false and non diagonal elements as true
~np.identity(4,dtype=bool)
```

```python
#access diagonal elements of numpy array
s=np.identity(4)
s.diagonal()
```

```python
~np.identity(4,dtype=int)
```

# **Numpy Slicing and indexing**

```python
x=np.arange(24).reshape((6,4))
x
```

```python
x[4]  #using index we can access one row at a time
```

```python
x[3][2] #row with element index
```

```python
x[4][3]
```

```python
x[2][2]
```

```python
x[2,1]
```

```python
x[5,2]
```

```python
x[2,3],[1,3]
```

```python
#replace 15 by 50
x[3,3]=50
x
```

```python
x[::-2]
```

```python
#access more rows
x[:]  #[start_index:stop_index]
```

```python
x[2:4,1:3]
```

```python
x[-3:]   #last 3 rows
```

```python
x[::-3]   #steping -3
```

```python
# column operations

x[-2:,-2:]   #[row,column]
```

```python
x[1:3,1:3]   #start row and stop    next start column stop column
```
