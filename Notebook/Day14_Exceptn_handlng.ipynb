# **Exception Handling**

```python
#Exception error
try:
 # print(30+20) # line without error
  print(30+'P')  # line with error
except:
  print('Error occur')

for i in range(5):
  print(i)
```

```python
# to handle the exception we must need 2 blocks
#try: to test the code
#except: to handle the exception
# to understand the error msg we need to use exception class
try:
  #print(30+'P')  # line with error
  print(30/0)
  #print(30+a)
except TypeError as msg:
  print('TypeError :occured',msg)
except ZeroDivisionError as msg:
  print('ZeroDivisionError :occured',msg)
except NameError as msg:
  print('NameError :occured',msg)
```

try: test code
except: handle the exception


else: if no exception then come here


finally: irrespective of above condition this block will execute

```python
try:
  #print(10/4) # without exception
  print(10/0) # with exception
except Exception as msg:
  print(msg)
else:
  print('No exception do below tasks....')
finally:
  print('Zukega nai sala....')
```

```python
try:
  balance = 5000
  amount = int(input('Enter amout in Rs.'))

  if  amount > balance:
    raise ValueError('Insufficient balance..')

except ValueError as e:
  print('Error:',e)

else:
  balance = balance - amount
  print('Withdrawal succesfull....')
  print('Remaining Amount is Rs. ',balance)

finally:
  print('Thank you for banking with us Visit Again.. :-)')
```
