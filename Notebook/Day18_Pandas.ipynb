```python
#how to concatanate 2 or more dataframes
a=pd.DataFrame([{'Name':'Aditi','Age':22,'Gender':'female'},
                {'Name':'Avanish','Age':15,'Gender':'Male'},
                {'Name':'Sayee','Age':16,'Gender':'Female'}])
b=pd.DataFrame([{'School':'Shantiniketan'},
 {'School':'Shantiniketan'},{'School':'Vimla Goenka'}])
c=pd.DataFrame([{'Roll no':201},{'Roll no':202},{'Roll no':203}])
d=pd.concat([a,b,c],axis=1)
d
```

```python
#how to perform merge operations
data=pd.DataFrame({'Names':['Aditi','Avanish','Sayee'],
                    'Roll no':[201,202,203]})
data2=pd.DataFrame({'Roll no':[201,202,203],
                     'School':['Shantiniketan','Shantiniketan','Vimla Goenka']})
final=pd.merge(data,data2,on='Roll no')
final
```
