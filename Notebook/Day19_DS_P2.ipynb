```python
#here in nunique we just get no. of unique names available in cat
cat_col=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for n in cat_col:
  print(n,':',c[n].nunique())
  print('='*30)
```

```python
#here it gives names of all unique cat available
cat_col=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for n in cat_col:
  print(n,':',c[n].unique())
```

```python
#in value_counts we get all vehicle types count separatly in decending order
cat_col=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for n in cat_col:
  print(n,':',c[n].value_counts())
```

```python
#if we want just 1st 3 op
cat_col=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type']

for n in cat_col:
  print(n,':',c[n].value_counts().head(3))
  print('='*30)
```
