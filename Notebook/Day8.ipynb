# **List comprehension**
-to create a new list in simple oneliner solution



-it uses the terneary operator to check if else option

```python
#syntax: [operation for var in sequence/iteration]
```

```python
d=[2,3,4,5,6,7]
#create a new list with squared elements
[num*num for num in d]
```

```python
nms=['prasad','imran','gayatri','jay','bhavana']
[nm.title() for nm in nms]
```

```python
[(nm.title(),len(nm)) for nm in nms]
```

```python
#interview question
#conditional list comprehension
#fetch those names whose len >5
#syntax:[operation for var in sequence/iteration if,condition]
#first element fetched in nm--->check condition len
[nm.title() for nm in nms if len(nm)>5]
```
