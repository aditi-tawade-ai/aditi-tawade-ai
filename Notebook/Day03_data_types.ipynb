**STRING METHODS**

```python
#Upper case - used to capitalize all the word
s= 'aditi tawade'
s.upper()
```

```python
#capitalize- makes only 1st letter captial
s.capitalize()
```

```python
#title- makes every 1st letter of each word capital
s.title()
```

```python
#lower- makes all words in lower case
s.lower()
```

```python
s='aDitI tAWAde'
s.title()
```

```python
#swapcase- lower to upper and upper to lower
s.swapcase()
```
