```python
# prepare data
# here we have probablity of 50% so below that its fail

marks = [[98],[80],[40],[20],[85],[10],[30]]  # input
result = ['Pass','Pass','Fail','Fail','Pass','Fail','Fail'] # output
```

```python
# For all ML operations we have one of the biggest Library
# Scikit-learn
# but we import it as sklearn
```

```python
# our problem is Classification problem
# import algorithm
from sklearn.linear_model import LogisticRegression

# LogisticRegression is an algorithm/model/class
```

```python
LogisticRegression
```

```python
# cretae an object of algor.
model = LogisticRegression()
model
```

```python
# Training--> fit() method
model.fit(marks,result)   #input,output
```

```python
model.classes_
```

```python
model.coef_
```

```python
#  Testing/Prediction--> predict()
model.predict([[15],[45],[99],[85]])
```

```python
model.predict([[89]])
```

```python
# to check the probablity use predict_proba()
model.predict_proba([[54]])
```

```python
model.classes_
```

```python
# Logistic Regression is probablistic model
model.classes_
```

```python
# internally it checks the probability
model.predict_proba([[58]])
```

```python
model.score(marks,result)*100
```
