# **Function with arguments**

```python
def addition(y,x):  #function
  print('Addition:',x+y)
  print('Multiplication:',x*y)
  print('Substraction:',x-y)
  print('division:',x%y)

addition(100,200)   #function calling
```

```python
#when we want to combine both we can do so
# but remember the rule
# first positional args then keywords

def sample(*args,**kwargs):
  print(args,kwargs)
sample(10,20,'aditi',num1=200,num2=500,num3=96)
```

```python
def marks(*n):
  print('marks obtained:',n)
marks(98,78,56,73,85,46,42,59,68,'absent')
```

```python
def marks(*m):
  for name,mark in m:
    print('Marks obtained by',name,':',mark)
marks(
    ('Aditi',98),
    ('Janvi',89),
    ('Sayee',78),
    ("Vedant",42),
    ('Pari','absent')
)
```
