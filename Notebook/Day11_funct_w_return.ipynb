# **Function with return keyword**

```python
# function will work on specific requirement
#it will generate one result
#this result we need to use as an input to the outside of function
#for this purpose we need to use return keyword

def test(balance):
  if balance <= 0:
    return 'No'
  else:
    return 'Yes'

test(120)
```

```python
def test(balance):
  if balance <= 0:
    return 'No'
  else:
    return 'Yes'

result = test(120)

if result == 'Yes':
  print('You can perform transaction')
else:
  print('You can\'t perform transaction')
  #hence when we want to bring the result of a function from inside to outside then
  #we need return to be used
```

```python
# purchase and billing example using return
items = {'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill = []
def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

# calling of purchase
total = purchase('wheat','milk')
print('Item amounts:',total)

# calling of check
ans = check(total)
print('Final bill amoutnt is Rs:',ans)
```
