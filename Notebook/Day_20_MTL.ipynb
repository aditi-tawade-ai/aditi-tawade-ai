# **Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
# We can perform different types Aof analysis
# 1. Univariate--> we are using single feature
# 2. Bi-Variate--> 2 features
# 3. Multi-variate--> More than 2 features (state vs population--> Gender)
```

```python
days=['mon','tue','wed','thur','fri']
rain=[120,30,25,20,80]
```

```python
# Bi-variate plots-line,scatter,bar,barh

plt.plot(days,rain)
```

```python
# chenge the colour of line to red
plt.plot(days,rain,'r')
```

```python
#change the line to --
plt.plot(days,rain,'g--')
```

```python
#give intersection point
plt.plot(days,rain,'g--*')
```

```python
plt.plot(rain,'g--*')
```

```python
plt.plot(days,rain,'m--*',label='Day wise rainfall in mm')
plt.grid()
plt.title('Weekly rainfall data')
plt.xlabel('Days Mon- Fri')
plt.ylabel('Rainfall in mm')
plt.legend()
```

```python
plt.plot(days,rain,'m--*',label='Day wise rainfall in mm')
#label='Day wise rainfall in mm' this gives plot name
plt.grid()
plt.title('Weekly rainfall data')
plt.xlabel('Days Mon- Fri')
plt.ylabel('Rainfall in mm')
plt.legend()  # prints label
plt.savefig('sample.png',dpi=100)  #save this fig
```

```python
plt.bar(days,rain,color= 'magenta',label='Daywise rainfall in mm')
plt.title('Weekly Rainfall data')
plt.xlabel('Days Mon- Fri')
plt.ylabel('Rainfall in mm')
plt.grid()
plt.legend()
```

```python
#if we want diff colors for each bar
plt.bar(days,rain,color= ['r','y','m','c','k'],label='Daywise rainfall in mm')
plt.title('Weekly Rainfall data')
plt.xlabel('Days Mon- Fri')
plt.ylabel('Rainfall in mm')
plt.grid()
plt.legend()
```

```python
#horizontal bar
plt.barh(days,rain,color= ['r','y','m','c','k'],label='Daywise rainfall in mm')
plt.title('Weekly Rainfall data')
plt.ylabel('Days Mon- Fri')
plt.xlabel('Rainfall in mm')
plt.grid()
plt.legend()
```

```python
import seaborn as sns
```

```python
#load datasets
print(sns.get_dataset_names())
```

```python
p=sns.load_dataset('penguins')
p.head(2)
```

```python
sns.countplot(data=p,x='island')
#data = p means read data stored in p and x is island names
# it automatically fetches unique island names and plot on graph with count
```

```python
p.columns
```

```python
sns.countplot(data=p,x='species',hue='sex',palette='Greens')

# hue is used to differentiate in the same column based on sub division
```

```python
sns.countplot(data=p,x='sex')
```
