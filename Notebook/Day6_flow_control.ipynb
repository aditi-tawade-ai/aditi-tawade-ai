# **Flow control block Introduction**
1. Selective / conditional statements
2. iterative statements
3. Transfer statements


```python
#1. selective satatements
#these statements are conditional based
#we used conditional operators to checck or compare objects
#conditional operators= <> <= >= == != these operators generate boolean answer means true or false
#in order to use these statements/blocks we have if elif else
```

```python
#when user wants to test single condition then use only if<condition>
```

```python
#if condition result is true then only program will enter inside the block
name='aditi'
if name=='aditi':
  print('Good morning')
```

```python
#if we want to check 2 conditions in one if block then use operators(and,or,not)
age=int(input('Enter your age:'))
country=input('Enter your country:')

if age> 18 and country==({'india','India','INDIA'}):
  print('you are eligible for vote')

else: #default when condition result is false
    print('you are not eligible for vote')
```

```python
#if for 1 check
#if else for 2 check
#if user wants to check more than 2 optios use elif
```

```python
#amt=120000
#amt=92000
#amt=55000
amt=35000

#whenever user takes a block in python it ends with : and at the next line
#your body of the block starts
#all statements we write inside the block. the space infront of each statement is called as
#indentation

if amt>100000:
  print('Purchase iphone')
elif amt>90000:
  print('Purchase samsung/vivo/realme phones')
elif amt> 50000:
  print('Purchase oppo/lava..etc phones')
else:
  print('You can try with remaining options..')
```

```python
# 2. iterative statements
# it contains 2 loops
# 1. is for loop
#2. while loop
```

```python
#for loop is an iterative loop which fetches one element at a time from
#the collection inside the body of element
a=[10,20,30,40]

#syntax for var in collection/iterable

for val in a:
  print(val)
```

```python
a=[10,20,30,40]
for val in a[::-1]:
  print(val)
```

```python
#take a dict and apply for loop
a={1:100,2:200,3:300,4:400}
type(a)
```

```python
#gives only key
for val in a:
  print(val)
```

```python
#gives both key and value
 for val in a:
  print(val,a[val])
```

```python
#converts in tuple
for val in a.items():
  print(val)
```

```python
num1,num2=(1,100) #tuple unpacking
num1
```

```python
num1,num2=(1,100)
num2
```

```python
for key,val in a.items():
  print(key,val)
```

```python
name='aditi'
for t in name:
  print(t)
```
